The learning algorithm used is based on vanilla Deep Q Learning including Experience Replay and Fixed Q-Targets techniques.
## Model Structure

```
self.fc1 = nn.Linear(37, 64)
self.fc2 = nn.Linear(64, 64)
self.fc3 = nn.Linear(64, 4)
```

- Replay buffer size 100,000
- Discount factor 0.99 (gamma)
- Soft update factor 0.001 (tau)
- Learning rate 0.0004 (alpha)
- Network update step interval 4
