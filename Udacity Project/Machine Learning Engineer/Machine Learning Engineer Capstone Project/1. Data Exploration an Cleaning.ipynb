```python
#import libraries and magic words
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
#define source directory
source_dir = 'source'
```

```python
#load in the data
azdias = pd.read_csv(os.path.join(source_dir, 'azdias.csv'))
customers = pd.read_csv(os.path.join(source_dir, 'customers.csv'))
```

Exploratory Data Analysis (EDA)
-----------------------------------------------

```python
#explore the data
display(azdias.head(), azdias.info(), azdias.describe())
display(customers.head(), customers.info(), customers.describe())
```

```python
#display 'object' columns
display(azdias.select_dtypes('object'),
        customers.select_dtypes('object'))
```

```python
#display three additional columns of customers.csv
display(customers['ONLINE_PURCHASE'].value_counts(),
      customers['PRODUCT_GROUP'].value_counts(),
      customers['CUSTOMER_GROUP'].value_counts())
```

```python
#make a copy to clean the data
azdias_clean = azdias.copy()
customers_clean = customers.copy()
```

```python
#count the 'object' column
azdias_clean['CAMEO_DEU_2015'].value_counts(dropna=False)
```

```python
#transform the 'object' columns to numeric columns

#create mapping dictionary
cameo_deu_dict = {'A': '1', 'B': '2', 'C': '3', 'D': '4', 'E': '5', 'F': '6', '0': '0'}

#fill NaN value and then transform to numeric column
azdias_clean['CAMEO_DEU_2015'] = azdias_clean['CAMEO_DEU_2015'].fillna('00')
azdias_clean['CAMEO_DEU_2015'] = azdias_clean['CAMEO_DEU_2015'].str.replace('XX', '00')
azdias_clean['CAMEO_DEU_2015'] = azdias_clean['CAMEO_DEU_2015'].str[:].map(lambda x: x.replace(x[1], cameo_deu_dict[x[1]]))
azdias_clean['CAMEO_DEU_2015'] = pd.to_numeric(azdias_clean['CAMEO_DEU_2015'])

customers_clean['CAMEO_DEU_2015'] = customers_clean['CAMEO_DEU_2015'].fillna('00')
customers_clean['CAMEO_DEU_2015'] = customers_clean['CAMEO_DEU_2015'].str.replace('XX', '00')
customers_clean['CAMEO_DEU_2015'] = customers_clean['CAMEO_DEU_2015'].str[:].map(lambda x: x.replace(x[1], cameo_deu_dict[x[1]]))
customers_clean['CAMEO_DEU_2015'] = pd.to_numeric(customers_clean['CAMEO_DEU_2015'])
```

```python
#count the 'object' column
azdias_clean['CAMEO_DEUG_2015'].value_counts(dropna=False)
```

```python
#fill NaN value and then transform to numeric column
azdias_clean.loc[azdias_clean['CAMEO_DEUG_2015'] == 'X', 'CAMEO_DEUG_2015'] = -1
azdias_clean['CAMEO_DEUG_2015'] = azdias_clean['CAMEO_DEUG_2015'].fillna(-1)
azdias_clean['CAMEO_DEUG_2015'] = pd.to_numeric(azdias_clean['CAMEO_DEUG_2015'])

customers_clean.loc[customers_clean['CAMEO_DEUG_2015'] == 'X', 'CAMEO_DEUG_2015'] = -1
customers_clean['CAMEO_DEUG_2015'] = customers_clean['CAMEO_DEUG_2015'].fillna(-1)
customers_clean['CAMEO_DEUG_2015'] = pd.to_numeric(customers_clean['CAMEO_DEUG_2015'])
```

```python
#count the 'object' column
azdias_clean['CAMEO_INTL_2015'].value_counts(dropna=False)
```

```python
#fill NaN value and then transform to numeric column
azdias_clean.loc[azdias_clean['CAMEO_INTL_2015'] == 'XX', 'CAMEO_INTL_2015'] = -1
azdias_clean['CAMEO_INTL_2015'] = azdias_clean['CAMEO_INTL_2015'].fillna(-1)
azdias_clean['CAMEO_INTL_2015'] = pd.to_numeric(azdias_clean['CAMEO_INTL_2015'])

customers_clean.loc[customers_clean['CAMEO_INTL_2015'] == 'XX', 'CAMEO_INTL_2015'] = -1
customers_clean['CAMEO_INTL_2015'] = customers_clean['CAMEO_INTL_2015'].fillna(-1)
customers_clean['CAMEO_INTL_2015'] = pd.to_numeric(customers_clean['CAMEO_INTL_2015'])
```

```python
#count the 'object' column
azdias_clean['D19_LETZTER_KAUF_BRANCHE'].value_counts(dropna=False)
```

```python
#drop the column
azdias_clean = azdias_clean.drop(['D19_LETZTER_KAUF_BRANCHE'], axis=1)
customers_clean = customers_clean.drop(['D19_LETZTER_KAUF_BRANCHE'], axis=1)
```

```python
#drop the column because EINGEFUEGT means insertion time which is meaningless
azdias_clean = azdias_clean.drop(['EINGEFUEGT_AM'], axis=1)
customers_clean = customers_clean.drop(['EINGEFUEGT_AM'], axis=1)
```

```python
#count the 'object' column
azdias_clean['OST_WEST_KZ'].value_counts(dropna=False)
```

```python
#fill NaN value and then transform to numeric column
azdias_clean['OST_WEST_KZ'] = azdias_clean['OST_WEST_KZ'].str.replace('W', '1')
azdias_clean['OST_WEST_KZ'] = azdias_clean['OST_WEST_KZ'].str.replace('O', '0')
azdias_clean['OST_WEST_KZ'] = azdias_clean['OST_WEST_KZ'].fillna('-1')
azdias_clean['OST_WEST_KZ'] = pd.to_numeric(azdias_clean['OST_WEST_KZ'])

customers_clean['OST_WEST_KZ'] = customers_clean['OST_WEST_KZ'].str.replace('W', '1')
customers_clean['OST_WEST_KZ'] = customers_clean['OST_WEST_KZ'].str.replace('O', '0')
customers_clean['OST_WEST_KZ'] = customers_clean['OST_WEST_KZ'].fillna('-1')
customers_clean['OST_WEST_KZ'] = pd.to_numeric(customers_clean['OST_WEST_KZ'])
```

```python
#count the 'object' column
customers_clean['PRODUCT_GROUP'].value_counts(dropna=False)
```

```python
#fill NaN value and then transform to numeric column
customers_clean['PRODUCT_GROUP'] = customers_clean['PRODUCT_GROUP'].str.replace('COSMETIC_AND_FOOD', '2')
customers_clean['PRODUCT_GROUP'] = customers_clean['PRODUCT_GROUP'].str.replace('FOOD', '1')
customers_clean['PRODUCT_GROUP'] = customers_clean['PRODUCT_GROUP'].str.replace('COSMETIC', '0')
customers_clean['PRODUCT_GROUP'] = pd.to_numeric(customers_clean['PRODUCT_GROUP'])
```

```python
#count the 'object' column
customers_clean['CUSTOMER_GROUP'].value_counts(dropna=False)
```

```python
#fill NaN value and then transform to numeric column
customers_clean['CUSTOMER_GROUP'] = customers_clean['CUSTOMER_GROUP'].str.replace('MULTI_BUYER', '1')
customers_clean['CUSTOMER_GROUP'] = customers_clean['CUSTOMER_GROUP'].str.replace('SINGLE_BUYER', '0')
customers_clean['CUSTOMER_GROUP'] = pd.to_numeric(customers_clean['CUSTOMER_GROUP'])
```

```python
#drop large NaN proportion columns, >33%
drop_columns = azdias_clean.columns[azdias_clean.isnull().sum() > azdias_clean.shape[0] / 3]
azdias_clean = azdias_clean.drop(drop_columns, axis=1)
customers_clean = customers_clean.drop(drop_columns, axis=1)
```

```python
#drop 'LNR' column
azdias_clean = azdias_clean.drop(['LNR'], axis=1)
customers_clean = customers_clean.drop(['LNR'], axis=1)
```

```python
#drop large NaN proportion rows, >33%
drop_azdias_rows = azdias_clean.index[azdias_clean.isnull().sum(axis=1) > azdias_clean.shape[1] / 3]
drop_customers_rows = customers_clean.index[customers_clean.isnull().sum(axis=1) > customers_clean.shape[1] / 3]

azdias_clean = azdias_clean.drop(drop_azdias_rows)
customers_clean = customers_clean.drop(drop_customers_rows)
```

```python
#reset dataframe index
azdias_clean = azdias_clean.reset_index(drop=True)
customers_clean = customers_clean.reset_index(drop=True)
```

```python
#fill Nan by 0
azdias_clean = azdias_clean.fillna(0)
customers_clean = customers_clean.fillna(0)
```

```python
#display the data to confirm all are cleaned
display(azdias_clean.head(), azdias_clean.info(), azdias_clean.describe())
display(customers_clean.head(), customers_clean.info(), customers_clean.describe())
```

Visualize the cleaned data
---------------------------

```python
#visualize the data
visualiza_list = ['CUSTOMER_GROUP', 'ONLINE_PURCHASE']
fig = plt.figure(figsize=(6, 10))

for i in range(len(visualiza_list)):
    ax = fig.add_subplot(len(visualiza_list), 1, i+1)
    ax = plt.bar(height = customers_clean[visualiza_list[i]].value_counts().values, x=['Others', 'Target'])
    plt.title(visualiza_list[i])
```

```python
#visualize the data
visualize_list = ['CAMEO_DEU_2015', 'CAMEO_DEUG_2015', 'CAMEO_INTL_2015']
fig = plt.figure(figsize=(12, 20))
n_bins = 10

for i in range(len(visualize_list)):
    ax = fig.add_subplot(len(visualize_list), 2, i+i+1)
    ax = plt.hist(azdias_clean[visualize_list[i]], bins=n_bins)
    title = '{} (General Population)'.format(visualize_list[i])
    plt.title(title, fontsize=12)
    
    ax2 = fig.add_subplot(len(visualize_list), 2, i+i+2)
    ax2 = plt.hist(customers_clean[visualize_list[i]], bins=n_bins)
    title = '{} (Customers)'.format(visualize_list[i])
    plt.title(title, fontsize=12)
```

Normalization and output the data
--------------------------------------------

```python
#scale numerical features into a normalized ragen, 0-1
scaler = MinMaxScaler()

azdias_scaled = pd.DataFrame(scaler.fit_transform(azdias_clean.astype(float)))
azdias_scaled.columns = azdias_clean.columns

customers_scaled = pd.DataFrame(scaler.fit_transform(customers_clean.astype(float)))
customers_scaled.columns = customers_clean.columns
```

```python
display(azdias_scaled.head(), azdias_scaled.info(), azdias_scaled.describe())
display(customers_scaled.head(), customers_scaled.info(), customers_scaled.describe())
```

```python
#save the data to local directory
azdias_scaled.to_csv(os.path.join(source_dir, 'azdias_scaled.csv'))
customers_scaled.to_csv(os.path.join(source_dir, 'customers_scaled.csv'))
```
