```python
#import libraries and magic words
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
#define source directory
source_dir = 'source'
```

```python
#load in the data
mailout_train = pd.read_csv(os.path.join(source_dir, 'mailout_train.csv'))
mailout_test = pd.read_csv(os.path.join(source_dir, 'mailout_test.csv'))
```

Exploratory Data Analysis (EDA)
-----------------------------------------------

```python
#explore the data
display(mailout_train.head(), mailout_train.info(), mailout_train.describe())
display(mailout_test.head(), mailout_test.info(), mailout_test.describe())
```

```python
#display 'object' columns
display(mailout_train.select_dtypes('object'),
        mailout_test.select_dtypes('object'))
```

```python
#make a copy to clean the data
mailout_train_clean = mailout_train.copy()
mailout_test_clean = mailout_test.copy()
```

```python
#transform the 'object' columns to numeric columns

#create mapping dictionary
cameo_deu_dict = {'A': '1', 'B': '2', 'C': '3', 'D': '4', 'E': '5', 'F': '6', '0': '0'}

#fill NaN value and then transform to numeric column
mailout_train_clean['CAMEO_DEU_2015'] = mailout_train_clean['CAMEO_DEU_2015'].fillna('00')
mailout_train_clean['CAMEO_DEU_2015'] = mailout_train_clean['CAMEO_DEU_2015'].str.replace('XX', '00')
mailout_train_clean['CAMEO_DEU_2015'] = mailout_train_clean['CAMEO_DEU_2015'].str[:].map(lambda x: x.replace(x[1], cameo_deu_dict[x[1]]))
mailout_train_clean['CAMEO_DEU_2015'] = pd.to_numeric(mailout_train_clean['CAMEO_DEU_2015'])

mailout_test_clean['CAMEO_DEU_2015'] = mailout_test_clean['CAMEO_DEU_2015'].fillna('00')
mailout_test_clean['CAMEO_DEU_2015'] = mailout_test_clean['CAMEO_DEU_2015'].str.replace('XX', '00')
mailout_test_clean['CAMEO_DEU_2015'] = mailout_test_clean['CAMEO_DEU_2015'].str[:].map(lambda x: x.replace(x[1], cameo_deu_dict[x[1]]))
mailout_test_clean['CAMEO_DEU_2015'] = pd.to_numeric(mailout_test_clean['CAMEO_DEU_2015'])
```

```python
#fill NaN value and then transform to numeric column
mailout_train_clean.loc[mailout_train_clean['CAMEO_DEUG_2015'] == 'X', 'CAMEO_DEUG_2015'] = -1
mailout_train_clean['CAMEO_DEUG_2015'] = mailout_train_clean['CAMEO_DEUG_2015'].fillna(-1)
mailout_train_clean['CAMEO_DEUG_2015'] = pd.to_numeric(mailout_train_clean['CAMEO_DEUG_2015'])

mailout_test_clean.loc[mailout_test_clean['CAMEO_DEUG_2015'] == 'X', 'CAMEO_DEUG_2015'] = -1
mailout_test_clean['CAMEO_DEUG_2015'] = mailout_test_clean['CAMEO_DEUG_2015'].fillna(-1)
mailout_test_clean['CAMEO_DEUG_2015'] = pd.to_numeric(mailout_test_clean['CAMEO_DEUG_2015'])
```

```python
#fill NaN value and then transform to numeric column
mailout_train_clean.loc[mailout_train_clean['CAMEO_INTL_2015'] == 'XX', 'CAMEO_INTL_2015'] = -1
mailout_train_clean['CAMEO_INTL_2015'] = mailout_train_clean['CAMEO_INTL_2015'].fillna(-1)
mailout_train_clean['CAMEO_INTL_2015'] = pd.to_numeric(mailout_train_clean['CAMEO_INTL_2015'])

mailout_test_clean.loc[mailout_test_clean['CAMEO_INTL_2015'] == 'XX', 'CAMEO_INTL_2015'] = -1
mailout_test_clean['CAMEO_INTL_2015'] = mailout_test_clean['CAMEO_INTL_2015'].fillna(-1)
mailout_test_clean['CAMEO_INTL_2015'] = pd.to_numeric(mailout_test_clean['CAMEO_INTL_2015'])
```

```python
#drop the column
mailout_train_clean = mailout_train_clean.drop(['D19_LETZTER_KAUF_BRANCHE'], axis=1)
mailout_test_clean = mailout_test_clean.drop(['D19_LETZTER_KAUF_BRANCHE'], axis=1)
```

```python
#drop the column because EINGEFUEGT means insertion time which is meaningless
mailout_train_clean = mailout_train_clean.drop(['EINGEFUEGT_AM'], axis=1)
mailout_test_clean = mailout_test_clean.drop(['EINGEFUEGT_AM'], axis=1)
```

```python
#fill NaN value and then transform to numeric column
mailout_train_clean['OST_WEST_KZ'] = mailout_train_clean['OST_WEST_KZ'].str.replace('W', '1')
mailout_train_clean['OST_WEST_KZ'] = mailout_train_clean['OST_WEST_KZ'].str.replace('O', '0')
mailout_train_clean['OST_WEST_KZ'] = mailout_train_clean['OST_WEST_KZ'].fillna('-1')
mailout_train_clean['OST_WEST_KZ'] = pd.to_numeric(mailout_train_clean['OST_WEST_KZ'])

mailout_test_clean['OST_WEST_KZ'] = mailout_test_clean['OST_WEST_KZ'].str.replace('W', '1')
mailout_test_clean['OST_WEST_KZ'] = mailout_test_clean['OST_WEST_KZ'].str.replace('O', '0')
mailout_test_clean['OST_WEST_KZ'] = mailout_test_clean['OST_WEST_KZ'].fillna('-1')
mailout_test_clean['OST_WEST_KZ'] = pd.to_numeric(mailout_test_clean['OST_WEST_KZ'])
```

```python
#drop large NaN proportion rows, >50%
drop_columns = mailout_train_clean.columns[mailout_train_clean.isnull().sum() > mailout_train_clean.shape[0] / 2]
mailout_train_clean = mailout_train_clean.drop(drop_columns, axis=1)
mailout_test_clean = mailout_test_clean.drop(drop_columns, axis=1)
```

```python
#drop 'LNR' column and save in a seperate dataframe
kaggle_lnr = pd.DataFrame(mailout_test_clean['LNR'])
mailout_train_clean = mailout_train_clean.drop(['LNR'], axis=1)
mailout_test_clean = mailout_test_clean.drop(['LNR'], axis=1)
```

```python
#drop large NaN proportion rows, >50%
drop_rows = mailout_train_clean.index[mailout_train_clean.isnull().sum(axis=1) > mailout_train_clean.shape[1] / 2]

mailout_train_clean = mailout_train_clean.drop(drop_rows)
```

```python
#fill Nan by 0
mailout_train_clean = mailout_train_clean.fillna(0)
mailout_test_clean = mailout_test_clean.fillna(0)
```

```python
#scale numerical features into a normalized ragen, 0-1
scaler = MinMaxScaler()

mailout_train_scaled = pd.DataFrame(scaler.fit_transform(mailout_train_clean.astype(float)))
mailout_train_scaled.columns = mailout_train_clean.columns

mailout_test_scaled = pd.DataFrame(scaler.fit_transform(mailout_test_clean.astype(float)))
mailout_test_scaled.columns = mailout_test_clean.columns
```

```python
#display the data to confirm all are cleaned
display(mailout_train_scaled.head(), mailout_train_scaled.info(), mailout_train_scaled.describe())
display(mailout_test_scaled.head(), mailout_test_scaled.info(), mailout_test_scaled.describe())
```

```python
#save the data to local directory
mailout_train_scaled.to_csv(os.path.join(source_dir, 'mailout_train_scaled.csv'))
mailout_test_scaled.to_csv(os.path.join(source_dir, 'mailout_test_scaled.csv'))
kaggle_lnr.to_csv(os.path.join(source_dir, 'kaggle_lnr.csv'))
```
